```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# box plot
```

```python
# median, quartile(25%, 50%, 75%), min, mix, outlier
```

```python
data= np.array([4,7,6,8,9,12,13,19,26,43,52,55,100])
plt.figure(figsize= (7,6))
sns.boxplot(y=data)
plt.title("Box Plot")
plt.grid(True)
plt.show()
```

```python
data= np.array([4,7,6,8,9,12,13,19,26,43,52,55,55,55,55,55,55,55,55,55,55,70])
plt.figure(figsize = (7,6))
sns.boxplot(y=data)
plt.title("Box Plot")
plt.grid(True)
plt.show()
```

```python
df= sns.load_dataset("titanic")
```

```python
df.head()
```

```python
plt.figure(figsize = (13,5))
plt.subplot(1,2,1)
sns.boxplot(x="class", y="age", data=df)
plt.title("Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")
plt.subplot(1,2,2)
sns.boxplot(x="class", y="fare", data=df)
plt.title("Class by Fare")
plt.xlabel("Class")
plt.ylabel("Fare")
plt.grid(True)
plt.show()
```

```python
sns.boxplot(x="sex", y="class", data=df)
plt.title("Sex By Survived")
plt.xlabel("Sex")
plt.ylabel("Survived")
plt.show()
```
