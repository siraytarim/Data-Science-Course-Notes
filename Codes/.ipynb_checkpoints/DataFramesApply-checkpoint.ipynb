```python
    def salary_category(salary):
        if 5000 > salary:
            return "Low"
        elif 5000 <= salary < 80000:
            return "Medium"
        else:
            return "High"
       
```

```python
df["Salary Category"] = df["Salary"].apply(salary_category)
```

```python
# 1. yol

def plus_performance_score(experience):
        if experience > 10:
            return 1
        else:
            return 0
```

```python
df["New_Score"] = df["Experience"].apply(plus_performance_score)
```

```python
df
```

```python
df["Adjusted"] = df["New_Score"] + df["Performance_Score"]
```

```python
# ikinci yol

def plus_performance(row):
        if row["Experience"] > 10:
            return row["Performance_Score"] + 1
        else:
            return row["Performance_Score"]
```

```python
df["Adjusted_Score"] = df.apply(plus_performance, axis = 1)
```
