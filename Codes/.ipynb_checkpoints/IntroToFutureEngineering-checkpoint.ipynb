```python
import seaborn as sns
import pandas as pd
```

```python
df= sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# imputation
```

# mean imputation

```python
sns.histplot(df.age, kde=True)
```

```python
df["Age mean"] = df.age.fillna(df.age.mean())
```

```python
df[["Age mean", "age"]]
```

```python
sns.boxplot(y =df["age"], data=df)
```

```python
# median
```

```python
df["age median"] = df.age.fillna(df.age.median())
```

```python
df
```

```python
df[["age median","age","Age mean"]]
```

```python
# mode imputation -> for categorical values
```

```python
df[df["embarked"].isnull()]
```

```python
df["embarked"].unique()
```

```python
mode_value = df[df["embarked"].notna()]["embarked"].mode()[0]
```

```python
df["embarked mode"] = df["embarked"].fillna(mode_value)
```

```python
df[["embarked mode", "embarked"]]
```
