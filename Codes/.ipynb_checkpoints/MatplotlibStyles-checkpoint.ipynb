```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.linspace(0,20,20)
data
```

```python
data2 = data **4
```

```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data, data2)
my_axes.plot(data2, data)
plt.show()
```

```python
new_fig, new_axes = plt.subplots()
new_axes.plot(data, data + 2, color="blue", linewidth = 1.0)
new_axes.plot(data, data + 6, color = "green", linewidth = 2.0)
new_axes.plot(data, data + 4, color = "green", linewidth = 2.0, linestyle= "--")
new_axes.plot(data, data + 5, color = "green", linewidth = 2.0, linestyle=":")
new_axes.plot(data, data + 7, color = "yellow", linewidth = 2.0, marker ="o", markersize = 5, markerfacecolor= "red")
plt.show()
```
