```python
# merge
```

```python
df_merge_1 = pd.read_csv("7-merge_data1.csv")
```

```python
df_merge_2 = pd.read_csv("7-merge_data2.csv")
```

```python
df_merge_1
```

```python
df_merge_2
```

```python
# merge - outer join
```

```python
df_merged_outer = pd.merge(df_merge_1, df_merge_2, on= "Employee_ID", how = "outer")
#bütün dataları ele alıp birleştirebildiklerini birleştiriyor
```

```python
df_merged_outer
```

```python
# merge - left join
```

```python
df_merged_left = pd.merge(df_merge_1, df_merge_2, on ="Employee_ID", how="left")
```

```python
df_merged_left
```

```python
# merge - right join
```

```python
df_merged_right = pd.merge(df_merge_1, df_merge_2, on ="Employee_ID", how = "right")
```

```python
df_merged_right
```

```python
# merge - inner join
```

```python
df_merged_inner = pd.merge(df_merge_1, df_merge_2, on ="Employee_ID", how = "inner")
```

```python
df_merged_inner
```
