```python
df_app_categories = df_clean.groupby(["Category", "App"])["Installs"].sum().reset_index()
```

```python
df_app_categories = df_app_categories.sort_values("Installs", ascending = False)
```

```python
apps = ["GAME", "COMMUNICATION", "TOOLS","PRODUCTIVITY","SOCIAL"]

plt.figure(figsize =(25,30))
for i, app in enumerate(apps):
    df2 = df_app_categories[df_app_categories.Category == app]
    df2 = df2.head(5)
    #print(df2)
    plt.subplot(3,2,i+1)
    sns.barplot(data=df2, x="Installs", y="App")
    plt.title(app, size = 10)
plt.tight_layout()
plt.show()
```
