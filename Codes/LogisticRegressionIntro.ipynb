```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("6-bank_customers.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df.drop("subscribed", axis = 1)
y = df["subscribed"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.3, random_state = 15)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
logistic.fit(X_train, y_train)
y_pred = logistic.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_pred, y_test)
print("score: ", score)
print(classification_report(y_pred,y_test))
print("Confusion matrix: \n", confusion_matrix(y_pred, y_test) )
```

```python
# hyperparameter tuning
```

```python
model = LogisticRegression()
```

```python
penalty = ["l1","l2","elasticnet"]
c_values = [100, 10, 1, 0.1, 0.01]
solver = ["newton-cg","lbfgs","liblinear","sag","saga","newton-cholesky"]
```

```python
params = dict(penalty = penalty, C = c_values, solver = solver)
```

```python
params
```

```python
# grid search cv
# verilen parametrelerin hepsini kombinasyon yaparak en iyi parametre koşullarını buluyor
```

```python
from sklearn.model_selection import GridSearchCV, StratifiedKFold
```

```python
cv = StratifiedKFold()
```

```python
grid = GridSearchCV(estimator = model, param_grid = params, cv=cv, scoring ="accuracy", n_jobs = -1)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
score = accuracy_score(y_pred, y_test)
print("score: ", score)
print(classification_report(y_pred,y_test))
print("Confusion matrix: \n", confusion_matrix(y_pred, y_test) )
```

```python
# random search cv
# parametreleri rastgele şekilde alıyor, kaç rastgele şey deneyeceğini biz ayarlıyoruz
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
model = LogisticRegression()
```

```python
randomcv = RandomizedSearchCV(estimator = model, param_distributions = params, cv = 5, n_iter = 20, scoring= "accuracy")
```

```python
randomcv.fit(X_train, y_train)
```

```python
randomcv.best_params_
```

```python
randomcv.best_score_
```

```python
y_pred = randomcv.predict(X_test)
```

```python
score = accuracy_score(y_pred, y_test)
print("score: ", score)
print(classification_report(y_pred,y_test))
print("Confusion matrix: \n", confusion_matrix(y_pred, y_test))
```
