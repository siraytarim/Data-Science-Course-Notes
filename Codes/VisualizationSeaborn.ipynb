```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
data = pd.read_csv("athlete_events.csv")
```

```python
data.describe()
```

```python
data.head()
```

```python
data.info()
```

```python
plt.scatter("Height","Weight", data=data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Height",y="Weight",data=data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Height",y="Weight", hue="Sex",data=data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.show()
```

```python
data.columns
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Height",y="Weight", hue="Sex", style="Medal",data=data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Height",y="Weight", hue="Sex", style="Medal", size="Age",data=data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Height",y="Weight", hue="Sex", size="Medal",data=data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x="Height",y="Weight",hue="Sex",data=data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x="Height",hue="Sex", data=data)
plt.ylabel("Frequency")
plt.title("Athlete Height Distribution")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.displot(x="Age",hue="Sex", data=data, kind="kde")
plt.ylabel("Frequency")
plt.title("Athlete Height Distribution")
plt.show()
```

```python
sns.barplot(x="Medal",y="Height", hue="Sex",data=data)
plt.title("Medals by Height")
plt.show()
```

```python
sns.barplot(x="Medal",y="Weight", hue="Sex",data=data)
plt.title("Medals by Weight")
plt.show()
```

```python
sns.catplot(x="Medal",y="Height", hue="Sex", col="Season",data=data)
plt.show()
# catplot numeric columnlarn korelasyonunu anlayabilmek iin nemli
```

```python
sns.heatmap(data.corr(numeric_only = True), annot=True)
```
