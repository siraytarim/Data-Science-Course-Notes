```python
columns = wq_data.columns
(fig, ax) = plt.subplots(4,4, figsize = (12,12))
ax=ax.flatten()

for i , column in enumerate(columns):
    sns.kdeplot(
        data = wq_data,
        x=column,
        hue = wq_data.quality,
        ax = ax[i]
    )
    ax[i].set_title(f"{column} Distribution")
    ax[i].set_xlabel(None)

for i in range(i+1, len(ax)):
    ax[i].axis("off")
```
